Input the width in terms of pixel count and desired color of border from the user
For both binary and greyscale images

```python
import cv2
import numpy as np
img = cv2.imread('../dataset/sample.jpg',0) # 0 for grayscale/binary image

border = int(input("Enter the width of border : "))

color = int(input("Enter the color of border : "))

padded_img = cv2.copyMakeBorder(img,border,border,border,border,cv2.BORDER_CONSTANT,value=color)

import matplotlib.pyplot as plt
plt.imshow(padded_img,cmap='gray')
plt.title('Padded Image')
plt.axis('off')
plt.show()
```
