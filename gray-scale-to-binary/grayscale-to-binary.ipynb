Greyscale to Binary

Case-1 : Use Mean Intensity as Threshold

```python
import cv2
import numpy as np
path = '../dataset/sample.jpg'
img = cv2.imread(path,cv2.IMREAD_GRAYSCALE) # read the image in grayscale
mean_value = np.mean(img)
throwaway, binary = cv2.threshold(img,mean_value,255,cv2.THRESH_BINARY)
print("Mean Intensity as Threshold : ",mean_value)

# cv2.imshow('Binary Image',binary)
# cv2.imshow('Original Image',img)


cv2.imwrite('./binary_mean.jpg',binary)
cv2.imwrite('./original_mean.jpg',img)
```

Case 2: User Input Threshold

```python
import cv2
import numpy as np
path = '../dataset/sample.jpg'
img = cv2.imread(path,cv2.IMREAD_GRAYSCALE) # read the image in grayscale
threshold = int(input("Enter the threshold value : "))
throwaway, binary = cv2.threshold(img,threshold,255,cv2.THRESH_BINARY)

cv2.imwrite('./binary_user.jpg',binary)
cv2.imwrite('./original_user.jpg',img)
```
