Case-1 : By taking mean average of three planes

```python
import cv2
import numpy as np
img = cv2.imread('../dataset/sample.jpg')

# split the image into three planes
b,g,r = cv2.split(img)
gray_mean = (b+g+r)//3

# showing using matplotlib
import matplotlib.pyplot as plt
plt.imshow(gray_mean,cmap='gray')
plt.title('Gray Image')
plt.axis('off')
plt.show()
```

Case 2: User enters weights for R, G, B (sum must be 1)

```python
import cv2
import numpy as np
img = cv2.imread('../dataset/sample.jpg')
b,g,r = cv2.split(img)

# taking weights from user
w1 = float(input("Enter weight for R: "))
w2 = float(input("Enter weight for G: "))
w3 = float(input("Enter weight for B: "))

if(w1+w2+w3 != 1):
    print("Invalid weights")
else:
    gray_weighted = w1*r + w2*g + w3*b

import matplotlib.pyplot as plt
plt.imshow(gray_weighted,cmap='gray')
plt.title('Gray Image')
plt.axis('off')
plt.show()
```
