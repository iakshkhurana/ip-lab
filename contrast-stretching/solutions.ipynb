```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("../dataset/sample.jpg", 0)

r1 = int(input("Enter input lower limit  : "))
r2 = int(input("Enter input upper limit  : "))
s1 = int(input("Enter output lower limit : "))
s2 = int(input("Enter output upper limit : "))

out = img.copy().astype(np.float32)

# Apply contrast stretching only in selected range
mask = (img >= r1) & (img <= r2)
out[mask] = ( (img[mask] - r1) / (r2 - r1) ) * (s2 - s1) + s1

# Clip + convert back to uint8
out = np.clip(out, 0, 255).astype(np.uint8)

# Show result
plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title("Original"); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(out, cmap='gray'); plt.title("Contrast Stretched"); plt.axis("off")
plt.show()
```
