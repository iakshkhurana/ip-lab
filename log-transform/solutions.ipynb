Q: Why do we normalize in log transform?

Because log transform produces small pixel values (0–5), not 0–255.
So we scale it back to 0–255, otherwise the image looks black.

Q: What is the purpose of log transform?

To brighten dark regions and enhance low intensity details.

```python
# assume c = 1
# s = c*log(1+r)
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('../dataset/sample.jpg',0)

c = 1
log_img = c*np.log(1+img)

log_img = np.uint8(255 * log_img / np.max(log_img))

plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title("Original"); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(log_img, cmap='gray'); plt.title("Log Transformed"); plt.axis('off')
plt.show()
```
