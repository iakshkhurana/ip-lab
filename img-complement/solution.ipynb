```python
import cv2
import numpy as np
img = cv2.imread('../dataset/sample.jpg',0)

gray_neg = 255 - img

_,binary_neg = cv2.threshold(gray_neg,127,255,cv2.THRESH_BINARY)

# cv2.threshold(src, thresh, maxval, type)

import matplotlib.pyplot as plt
plt.subplot(1,3,1); plt.imshow(img, cmap='gray'); plt.title("Original"); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(gray_neg, cmap='gray'); plt.title("Grayscale Neg"); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(binary_neg, cmap='gray'); plt.title("Binary Neg"); plt.axis('off')
plt.show()
```
