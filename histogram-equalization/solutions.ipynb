```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("../dataset/sample.jpg", 0)

eq_img = cv2.equalizeHist(img)

plt.subplot(1,2,1); 
plt.hist(img.ravel(),256,[0,256]); 
plt.title("Original"); 
plt.axis('off')
plt.subplot(1,2,2); 
plt.hist(eq_img.ravel(),256,[0,256]); 
plt.title("Histogram Equalized"); 
plt.axis('off')
plt.show()
```
